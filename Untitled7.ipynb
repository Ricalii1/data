```python
#перевод валюты (из руб в дол) Был создан дополнительный атрибут, который демонстрирует курс рубля к юаням.
df3['INCOME_MAIN_AMT_EUR'] = df3['INCOME_MAIN_AMT'].apply(lambda x: x / 80)
df3['INCOME_MAIN_AMT_CNY'] = df3['INCOME_MAIN_AMT'].apply(lambda x: x / 11)
df3
```

```python
#1.5
X = df3[["AGE", "GENDER", "EDUCATION_LEVEL"]]
#атрибуты для визуализации
#в зависимости от региона и категории товара (продовца) стоимость покупок будет разной
```

```python
#Обьявляю метод главных компонент
pca = PCA(n_components=2)
#Применяю его на данных
X = pca.fit_transform(X)
```

```python
#обьявляю MinMaxScaler
scaler = MinMaxScaler()
#преобразую данные
X = scaler.fit_transform(X)
```

```python
#Функция для визуализации распределения 
def viz(prediction):
    #Размер фигуры
    plt.figure(figsize=(12, 12))
    plt.subplot(224)
    #Выводить изображение буду при помощи scatter
    plt.scatter(X[:, 0], X[:, 1], c=prediction)
    plt.title("Unevenly Sized Blobs")
    #Вывод изображения
    plt.show()
```

```python
#Настройка параметров
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
#Предсказание
kmpreds = kmeans.predict(X)
#Заношу кластеризированные метки в набор данных
df3["KMCLUSTS"] = kmpreds

#Применяю функцию описаную выше
viz(df3["KMCLUSTS"])
```

```python
#Настройка параметров
kbmeans = MiniBatchKMeans(n_clusters=4, random_state=0, batch_size=4096)
#Заношу кластеризированные метки в набор данных
df3["KBCLUSTS"] = kbmeans.fit_predict(X)

viz(df3["KBCLUSTS"])
```

```python
#Настройка параметров
gm = GaussianMixture(n_components=2, random_state=0).fit_predict(X)
#Заношу кластеризированные метки в набор данных
tk_merge["GMCLUSTS"] = gm

viz(tk_merge["GMCLUSTS"])
```

```python
#Использую метрику calinski_harabasz_score
print(sklearn.metrics.calinski_harabasz_score(X, df3["KMCLUSTS"]))
#Использую метрику davies_bouldin_score
print(sklearn.metrics.davies_bouldin_score(X, df3["KMCLUSTS"]))
```
